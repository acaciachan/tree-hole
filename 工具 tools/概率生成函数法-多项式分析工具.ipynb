```python
import numpy as np
from fractions import Fraction


def analyze_polynomial(p):
    # 获取多项式的系数和次数
    coefficients = p.coefficients
    degrees = np.arange(len(coefficients) - 1, -1, -1)

    # 计算所有系数的总和（分母）
    total = np.sum(coefficients)

    # 打印表头
    print("| 次数 | 权重 | 分数概率 | 百分比概率 |")
    print("|------|------|----------|------------|")

    # 遍历每一项，计算概率并打印
    for degree, coeff in zip(degrees, coefficients):
        if coeff == 0:
            continue  # 忽略系数为0的项
        fraction_prob = f"{coeff}/{total}"  # 分数概率
        percent_prob = (coeff / total) * 100  # 百分比概率
        print(f"| {degree:4} | {coeff:4} | {fraction_prob:8} | {percent_prob:10.2f}% |")

    # 计算期望值
    expectation = np.sum(degrees * (coefficients / total))
    expectation_fraction = Fraction(int(expectation * total), total)  # 分数形式

    # 打印期望值
    print("\n期望值（小数形式）:", expectation)
    print("期望值（分数形式）:", expectation_fraction)


# 输入多项式
p = np.poly1d([1, 1, 1, 1, 0, 0]) ** 3

# 输出多项式分析
analyze_polynomial(p)
```
